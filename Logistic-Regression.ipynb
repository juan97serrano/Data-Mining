<h1 align = "center" style="background-color:powderblue;"> Importamos los Datos</h1>

```python
Banco = pd.read_csv("P3/bank.csv",sep=';')
```

<h1 style="background-color:powderblue;"> EJERCICIO 1: Prepocesamiento</h1>

### Descripción del DataSet

```python
Banco.describe()
```

### Si hay valores con NaN los eliminamos

```python
Banco = Banco.dropna()
Banco
```

### Categorizamos todas  las columnas del DataFrame

```python
lista =['job','marital','education','default','housing','loan','contact','month','poutcome','y']

for lista2 in lista:

    Banco[[lista2]]=Banco[[lista2]].apply(lambda x: pd.factorize(x)[0])
```

### Seleccionamos los datos para el entrenamiento

```python
X = np.array(Banco.drop(['y'],1))
y = np.array(Banco['y'])
```

<h1 style="background-color:powderblue;">EJERCICIO 2: Construimos el modelo</h1>

### Utilizamos la regresión logística ya que es la regresión adecauda para predecir una variable categórica.

```python
model = linear_model.LogisticRegression()
model.fit(X,y)
```

### Vemos las variables más significativas

```python
model.coef_
```

### Estandarizamos los valores del DataFrame

```python
model.score(X,y)
```

### Interpretación del modelo

Analizando los coeficientes obtenidos vemos que  las variables que mas influyen son la de la posición 7 y 8. Que corresponden con housing y loan.
