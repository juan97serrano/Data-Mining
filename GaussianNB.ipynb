<h1 align = "center" style="background-color:powderblue;"> Importamos las librerias</h1>


```python
%reset
import pandas as pd
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn import metrics
from sklearn import datasets, svm, metrics
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt 

from sklearn.naive_bayes import GaussianNB
```

<h1 align = "center" style="background-color:powderblue;"> Importamos los datos</h1>


```python
Datos = pd.read_csv("lluvias.csv")
Datos
```

<h1 style="background-color:powderblue;"> EJERCICO 1</h1>

<h1 align = "center" style="background-color:powderblue;"> Preprocesamiento de los datos</h1>


### Cambiamos la temperatura a grados centigrados

```python
Datos['TMAX'] = (Datos['TMAX']-32)*(5/9)
Datos['TMIN'] = (Datos['TMIN']-32)*(5/9)
Datos
```

### Categorizamos las columnas TMIN y TMAX

Se considera que la temperatura es baja si está por debajo de 10 grados, media si está entre 10 y 20 y alta si está por encima de 20. 

```python
Baja=1
Media=2
Alta =3

Datos['TMIN'] = pd.cut(Datos['TMIN'], bins=[float("-inf"), 10, 20, float("inf")], labels=[Baja,Media,Alta])
Datos['TMAX'] = pd.cut(Datos['TMAX'], bins=[float("-inf"), 10, 20, float("inf")], labels=[Baja,Media,Alta])
Datos
```

### Categorizamos la columna PRCP

Se considera que ha llovido poco si la medida está por debajo de 1, normal si está entre 1 y 2, y que ha llovido mucho si está por encima de 2.

```python
Poco=1
Normal=2
Mucho =3

Datos['PRCP'] = pd.cut(Datos['PRCP'], bins=[0, 1, 2, float("inf")], include_lowest = True, labels=[Poco,Normal,Mucho])
```

### Categorizamos la columna RAIN

```python
Datos['RAIN'] = Datos.RAIN.astype('category').cat.codes.astype(int)
```

### Quitamos los NAN

```python
Datos = Datos.dropna()
```

# Resultado datos preprocesados

```python
Datos
```

Preparamos datos para el entrenamiento y la clasificación

```python
X = Datos.drop(['RAIN','DATE'], axis=1) 
Y = Datos['RAIN']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.20,shuffle= False) 
```

<h1 style="background-color:powderblue;"> EJERCICO 2</h1>

<h1 align = "center" style="background-color:powderblue;"> Crea un clasificador e indica su error</h1>

La validación del modelo se podría realizar con score, pero hemos preferido realizarlo nosotros manualmente prediciendo y comparando con el resultado esperado. 

```python
# Construimos el clasificador
classifier = GaussianNB()

# Entrenamos el clasificador
classifier.fit(X_train, y_train)

# Predecimos 
predicted = classifier.predict(X_test)

# Las siguientes matrices nos indican el nivel de precisión de la predicción
print("Classification report for classifier %s:\n%s\n"
      % (classifier, metrics.classification_report(y_test, predicted)))
print("Confusion matrix:\n%s" % metrics.confusion_matrix(y_test, predicted))
```
